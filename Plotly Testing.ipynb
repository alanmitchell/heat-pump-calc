````python
def convert_co2_to_miles_driven(co2_saved):
    ''' Converts CO2 emissions to a mileage driven
    equivalent for vehicles in the U.S. using EPA
    methodology:  https://www.epa.gov/energy/greenhouse-gases-equivalencies-calculator-calculations-and-references#miles'''
    pounds_in_metric_ton = 2204.62
    tons_co2_per_gallon = 0.0089
    avg_gas_mileage_us_fleet = 22
    mileage_equivalent = co2_saved / pounds_in_metric_ton / tons_co2_per_gallon * avg_gas_mileage_us_fleet
    
    return mileage_equivalent
````

````python
cash_flow = calc.cash_flow()
````

````python
# Loweer the initial cash flow to see if the graph works
cash_flow.set_value(0, 'cash_flow', -2000)
````

````python
cash_flow['cumulative_cash_flow'] = cash_flow.get_value(0, 'cash_flow')

for index, row in cash_flow.iterrows():
    
    if index == 0:
        pass
    
    else:
        cumulative_cash_flow = cash_flow.get_value((index-1), 'cumulative_cash_flow')
        current_cash_flow = cash_flow.get_value(index, 'cash_flow')
        cash_flow.set_value(index, 'cumulative_cash_flow', (cumulative_cash_flow + current_cash_flow))

cash_flow

    
    
````

````python
cash_flow['energy_cost_difference'] = cash_flow.fuel_cost_savings - cash_flow.elec_cost_increase
cash_flow['energy_cost_increase'] = np.where(cash_flow.energy_cost_difference <= 0,
                                            cash_flow.energy_cost_difference,
                                            0)
cash_flow['energy_cost_decrease'] = np.where(cash_flow.energy_cost_difference > 0,
                                            cash_flow.energy_cost_difference,
                                            0)

# Split cash flow into positive and negative to fill areas with different colors in chart
cash_flow['negative_cash_flow'] = np.where(cash_flow.cumulative_cash_flow <= 0, 
                                          cash_flow.cumulative_cash_flow,
                                          0)
cash_flow['positive_cash_flow'] = np.where(cash_flow.cumulative_cash_flow >= 0,
                                          cash_flow.cumulative_cash_flow,
                                          0)

cash_flow
````

````python
negative_cash_flow = go.Scatter(x=cash_flow.index,
                            y=cash_flow.negative_cash_flow,
                            name='Cash Flow ($)',
                            fill='tozeroy',
                            fillcolor='#d7191c',
                            line=dict(color='#ffffff')
                           )

positive_cash_flow = go.Scatter(x=cash_flow.index,
                            y=cash_flow.positive_cash_flow,
                            name='Cash Flow ($)',
                            fill='tozeroy',
                            fillcolor='#000000',
                            line=dict(color='#ffffff')
                           )

data = [negative_cash_flow, positive_cash_flow]

layout = go.Layout(title='Heat Pump Lifetime Cash Flow',
                   xaxis=dict(title='Month'),
                   yaxis=dict(title='Annual Cash Flow', hoverformat='.0f'),
                  )

fig = go.Figure(data=data, layout=layout)

py.iplot(fig, filename='heatpump_cash_flow', fileopt='overwrite')
````
